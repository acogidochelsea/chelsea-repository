```python
import requests
import time
import matplotlib.pyplot as plt
```

```python
latitudes = []
longitudes = []
timestamps = []

for i in range(100):
    try:
        response = requests.get("https://api.wheretheiss.at/v1/satellites/25544", timeout=5)
        data = response.json()

    
        if "latitude" in data and "longitude" in data and "timestamp" in data:
            latitudes.append(data["latitude"])
            longitudes.append(data["longitude"])
            timestamps.append(data["timestamp"])
            print(f"{i+1}: {data['latitude']}, {data['longitude']}")
        else:
            print(f"{i+1}: Error with coordinate data")
    except Exception as e:
        print(f"Error at point {i+1}: {e}")
    
    time.sleep(10)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(longitudes, latitudes, marker='o', linestyle='-', color='blue')
plt.title('Path of the International Space Station')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```
